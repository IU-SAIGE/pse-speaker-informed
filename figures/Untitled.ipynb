```python
import numpy as np
import matplotlib.pyplot as plt
import os, pathlib
import json, yaml
import librosa

import sys
sys.path.append('/media/sdb1/asivara/Research/2021_waspaa/')
import _datasets as D
```

```python
from scipy.io import wavfile

sr, s = wavfile.read(D.librispeech.filepath.iloc[0])
s = s / (abs(s).max())
S = librosa.stft(s, n_fft=1024, hop_length=256)
Sm = np.abs(S)
```

```python
# waveform
fig, ax = plt.subplots(figsize=(0.5, 0.2), dpi=300)
plt.plot(s, c=plt.cm.tab10(0), lw=0.2)
plt.axis('off')
plt.grid('off')
plt.savefig('fig_waveform.pdf', bbox_inches='tight', facecolor='white', pad_inches=0)
plt.show()
```

```python
# waveform
fig, ax = plt.subplots(figsize=(0.5, 0.2), dpi=300)
plt.imshow(np.log(Sm+1e-2), aspect='auto', origin='lower', cmap=plt.cm.inferno)
plt.axis('off')
plt.grid('off')
plt.savefig('fig_spectrogram.pdf', bbox_inches='tight', facecolor='white', pad_inches=0)
plt.show()
```
